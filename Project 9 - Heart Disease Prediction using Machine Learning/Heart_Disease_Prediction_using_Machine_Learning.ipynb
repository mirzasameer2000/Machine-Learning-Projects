```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
heart_dataset = pd.read_csv('/content/heart_disease_data.csv')
```

```python
# 1 = having disease,  0 = not having disease
heart_dataset.head()
```

```python
heart_dataset.shape
```

```python
heart_dataset.info()
```

```python
heart_dataset.isnull().sum()
```

```python
heart_dataset.describe()
```

```python
heart_dataset['target'].value_counts()
```

- Defective Heart = 1
- Healthy Heart = 0

```python
X = heart_dataset.drop(columns='target', axis=1)
Y = heart_dataset['target']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Model is overfitting

```python
cm = confusion_matrix(X_test_prediction, Y_test)
print('classification_report:' , cm)
```

```python
classification_report = classification_report(X_test_prediction, Y_test)
print('classification_report:' , classification_report)
```

```python
labels = ['Training Accuracy', 'Test Accuracy']
accuracy_scores = [training_data_accuracy, test_data_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(labels, accuracy_scores, color=['skyblue', 'lightcoral'])
plt.ylim(0, 1) # Accuracy ranges from 0 to 1
plt.ylabel('Accuracy')
plt.title('Comparison of Training and Test Accuracy')
plt.show()
```
